## Selecting Series data

```python
college = pd.read_csv('master/data/college.csv', index_col='INSTNM')
city = college['CITY']
city.head()
```

```python
# Single selection return one object, here a string:
city.iloc[3]
```

```python
# Multiple selection through a list return a Series
# This feature is an enhancement over a Python list, which is incapable of selecting multiple disjoint items in this manner.
city.iloc[[10,20,30]]
```

```python
# Slice notation, to select equally spaced partition of data, fonctionne comme suit:
# city.iloc[debut (compris) : fin (non compris) : espacement]

# ici nous aurons les index 4, 14, 24, 34 et 44 de retourn√©s

city.iloc[4:50:10]
```

```python
city.loc['Heritage Christian University']
```

```python
city.index
```

```python
np.random.seed(1)
labels = list(np.random.choice(city.index, 4))
labels
```

```python
city.loc[labels]
```

```python
# La borne de fin est comprise contrairement a iloc:
city.loc['Alabama State University':'Reid State Technical College':10]
```

```python
# if we want to retain the Series type return when we pass a scalar value:
city.iloc[[3]]

# SCALAR VALUE: a integer, a floating number or a string.
```

```python
# Beware of order, if we switch start and stop, an empty Series will be returned:
city.loc['Reid State Technical College':'Alabama State University':10]
```

## Speeding up scalar selection

```python
college = pd.read_csv('master/data/college.csv', index_col='INSTNM')
cn = 'Texas A & M University-College Station'
college.loc[cn, 'UGDS_WHITE']
```

```python
college.at[cn, 'UGDS_WHITE']
```

```python
%timeit college.loc[cn, 'UGDS_WHITE']
```

```python
%timeit college.at[cn, 'UGDS_WHITE']
```

```python
row_num = college.index.get_loc(cn)
col_num = college.columns.get_loc('UGDS_WHITE')
row_num, col_num
```

```python
%timeit college.iloc[row_num, col_num]
```

```python
%timeit college.iat[row_num, col_num]
```

```python
state = college['STABBR']
state.iat[1000]
```

```python
state.at['Stanford University']
```

## Slicing lexicographically

```python
college = pd.read_csv('master/data/college.csv', index_col='INSTNM')
```

```python
# college.loc['Sp':'Su']
```

```python
college = college.sort_index()
college.head()
```

```python
#.loc raises KeyError when labels are not found. Special exception when the index is lexicographically sorted,
# and a slice is passed to it even if values are not the exact values of the index.

college.loc['Sp':'Su']
```

```python
college = college.sort_index(ascending=False)
college.index.is_monotonic_decreasing
```

```python
college.loc['E':'B']
# Python sorts all capital letters before lowercase and all integers before capital letters.
```
