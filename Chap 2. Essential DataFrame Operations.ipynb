## Chaining DataFrame methods together

```python
movie = pd.read_csv('master/data/movie.csv')

# The isnull method returns a DataFrame the same size as the calling DataFrame 
# but with all values transformed to booleans.

movie.isnull().head()
```

```python
# We have only booleans in this DF. See the counts of the following data types to verify this:
movie.isnull().get_dtype_counts()
```

```python
movie.isnull().sum().head()
```

```python
movie.isnull().sum().sum()
```

```python
movie.isnull().any().any()
```

```python
movie[['color', 'movie_title', 'color']].max()
```

```python
movie.select_dtypes(['object']).fillna('').min()
```

```python
# rewrite the above chain on multiple lines for better readibility
movie.select_dtypes(['object']) \
     .fillna('') \
     .min()
```

## Working with operators on a DataFrame

```python
college = pd.read_csv('master/data/college.csv')
```

```python
college.head()
```

```python
# college + 5
# TypeError: Could not operate 5 with block values can only concatenate str (not "int") to str
```

```python
college = pd.read_csv('master/data/college.csv', index_col='INSTNM')
```

```python
college_ugds_ = college.filter(like='UGDS_')
```

```python
college_ugds_.head()
```

```python
college_ugds_ + .00501
```

```python
(college_ugds_ + .00501) // .01
```

```python
college_ugds_op_round = (college_ugds_ + .00501) // .01 / 100
```

```python
college_ugds_op_round.head()
```

```python
college_ugds_round = (college_ugds_ + .00001).round(2)
```

```python
college_ugds_op_round.equals(college_ugds_round)
```

```python
college_ugds_op_round_methods = college_ugds_.add(.00501) \
                                             .floordiv(.01) \
                                             .div(100)
```

```python
college_ugds_op_round_methods.equals(college_ugds_op_round)
```
